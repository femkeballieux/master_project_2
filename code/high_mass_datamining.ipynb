```python
#This one is useful to test stuff, if DALquery error, server is out. use either ESO, NRAO or NAOJ
myquery = alminer.target(["AB Aur"] , tap_service='NAOJ')
```

```python
#The high mass query
high_mass_query = alminer.keysearch({'proposal_abstract': ['"high-mass star formation" outflow',
                                    '"massive star formation" outflow', '"high-mass outflow" "star formation"', 
                                    '"massive outflow" "star formation"'], 
                                    'scientific_category':['ISM and star formation']},
                                    print_targets=True, tap_service='NAOJ')
```

```python
#This step takes out uranus.... So no water vapour measurements but only the science data
filtered_high_mass = high_mass_query[high_mass_query['scan_intent']=="TARGET"]
alminer.summary(filtered_high_mass)

#TODO: check all other things we could filter on
```

```python
#Here we filter on only the CO observations
high_mass_CO_abstract = alminer.CO_lines(filtered_high_mass, print_targets=True)
```

```python
#Here we download the overview of the whole query
alminer.save_table(filtered_high_mass, filename="/net/vdesk/data2/bach1/ballieux/master_project_2/tables/high_mass_filtered")

#This filters it on only the CO lines. Sorts it on the different lines, some sources are thus present multiple times for different lines. 
alminer.save_table(high_mass_CO_abstract, filename="/net/vdesk/data2/bach1/ballieux/master_project_2/tables/high_mass_filtered_CO")
```
